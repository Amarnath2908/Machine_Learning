```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
```

```python
x=df[["pickup_latitude","pickup_longitude"]]
```

```python
s = StandardScaler()
x_scaled = s.fit_transform(x)
```

```python
sample_size = 50000 

if len(x_scaled) > sample_size:
    idx = np.random.choice(len(x_scaled), sample_size, replace=False)
    x_sample = x_scaled[idx]
    x_original_sample = x.iloc[idx]   # if x is a dataframe
else:
    x_sample = x_scaled
    x_original_sample = x

eps_values = [0.2, 0.3, 0.5]

for eps in eps_values:
    print(f"\nEvaluating DBSCAN using eps={eps}")

    dbscan = DBSCAN(eps=eps, min_samples=5)
    labels = dbscan.fit_predict(x_sample)

    unique_labels = set(labels)
    n_clusters = len(unique_labels) - (1 if -1 in labels else 0)
    n_noise = list(labels).count(-1)
    noise_ratio = n_noise / len(labels)

    print(f"Number of clusters: {n_clusters}")
    print(f"Noise ratio: ", round(noise_ratio, 2))

    if n_clusters > 1:
        x_non_noise = x_sample[labels != -1]
        labels_non_noise = labels[labels != -1]

        sc = StandardScaler()
        x_scaled_non_noise = sc.fit_transform(x_non_noise)

        score = silhouette_score(x_scaled_non_noise, labels_non_noise)
        print("Silhouette Score:", round(score, 3))
```
