```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
x = df[["enginesize"]]
y = df["price"]
```

```python
lin_pred = LinearRegression()
```

```python
lin_pred.fit(x,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_reg = PolynomialFeatures(degree = 4)
x_poly = poly_reg.fit_transform(x)
lin_regx = LinearRegression()
lin_regx.fit(x_poly,y)
```

```python
x_sorted = x.sort_values(by = "enginesize")
x_poly_sorted = poly_reg.transform(x_sorted)
y_poly_pred = lin_regx.predict(x_poly_sorted)
```

```python
plt.scatter(x,y,color = 'red')
plt.plot(x,lin_pred.predict(x),color = 'blue')
plt.plot(x,y_poly_pred,color = 'blue')
plt.xlabel("Engine-size")
plt.ylabel("Price")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures


X = df[['enginesize']]
y = df['price']

# -----------------------------
# 1) Train Linear Regression
# -----------------------------
lin_model = LinearRegression()
lin_model.fit(X, y)

# -----------------------------
# 2) Train Polynomial Regression (degree 2)
# -----------------------------
poly = PolynomialFeatures(degree=4)
X_poly = poly.fit_transform(X)

poly_model = LinearRegression()
poly_model.fit(X_poly, y)

# -----------------------------
# 3) Create smooth X range for curve
# -----------------------------
X_range = np.linspace(X.min().values[0], X.max().values[0], 300).reshape(-1, 1)

# Linear predictions
y_lin_pred = lin_model.predict(X_range)

# Polynomial predictions
X_range_poly = poly.transform(X_range)
y_poly_pred = poly_model.predict(X_range_poly)

# -----------------------------
# 4) Plot everything
# -----------------------------
plt.figure(figsize=(10,6))

# Actual data
plt.scatter(X, y, alpha=0.7, label="Actual Data Points")

# Linear line
plt.plot(X_range, y_lin_pred, linewidth=2, label="Linear Regression Fit")

# Polynomial curve
plt.plot(X_range, y_poly_pred, linewidth=2, label="Polynomial Regression (Degree 4) Fit")

plt.xlabel("Engine Size")
plt.ylabel("Car Price")
plt.title("Engine Size vs Car Price with Model Fits")
plt.legend()
plt.grid(True)

```
