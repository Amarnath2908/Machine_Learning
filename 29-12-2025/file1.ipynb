From the above 2 graphs(histplot and barplot), Barplot : we can conclude that yellow color receieves higher tips compared to green color.The yellow color is more spread out compared to the green color. Histplot : yellow color tips are more concentrated at higher tips even though they are significantly small whereas green color is concentrated at lower tips.

H0: distribution of tip for yellow = distribution of tip for green,

H1: distribution of tip for yellow != distribution of tip for green

```python
yellow_tip = df["tip"].loc[df["color"]=="yellow"]
green_tip = df["tip"].loc[df["color"] == "green"]
```

```python
green_tip,yellow_tip
```

```python
green_tip.shape, yellow_tip.shape
```

```python
# Independent T-Test

from scipy import stats
t_stat, p_val = stats.ttest_ind(yellow_tip, green_tip)
print("Independent T-Test statistic:", t_stat)
print("P-value:", p_val)
```

```python

if p_val < 0.05:
    print("Reject null hypothesis:")
else:
    print("Fail to reject null hypothesis: ")
```

H0: Mean tip is same across all pickup_borough categories,

H1: Atleast one category has different distribution of tip

```python
m_tip=df["tip"].loc[df["pickup_borough"] == "Manhattan"]
b_tip=df["tip"].loc[df["pickup_borough"] == "Bronx"]
q_tip=df["tip"].loc[df["pickup_borough"] == "Queens"]
br_tip=df["tip"].loc[df["pickup_borough"] == "Brooklyn"]
```

```python
# ANOVA Test
f_stat, p_val = stats.f_oneway(m_tip,b_tip,q_tip,br_tip)
print("ANOVA F-statistic:", f_stat)
print("P-value:", p_val)

if p_val < 0.05:
    print("Reject null hypothesis:")
else:
    print("Fail to reject null hypothesis:")

```

- H0: Distribution of group1 = Distribution of group2  
- H1: Distribution of group1 ≠ Distribution of group2
