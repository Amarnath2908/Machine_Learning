```python
pickupf_tip = df["tip"].loc[df["pickup_month"]=="February"]
pickupm_tip = df["tip"].loc[df["pickup_month"] == "March"]
```

```python
# Independent T-Test

from scipy import stats
t_stat, p_val = stats.ttest_ind(pickupf_tip, pickupm_tip)
print("Independent T-Test statistic:", t_stat)
print("P-value:", p_val)
```

```python

if p_val < 0.05:
    print("Reject null hypothesis:")
else:
    print("Fail to reject null hypothesis: ")
```

```python
pum_tip = df["tip"].loc[df["pickup_dayname"] == "Monday"]
put_tip = df["tip"].loc[df["pickup_dayname"] == "Tuesday"]
puw_tip = df["tip"].loc[df["pickup_dayname"] == "Wednesday"]
puth_tip = df["tip"].loc[df["pickup_dayname"] == "Thursday"]
puf_tip = df["tip"].loc[df["pickup_dayname"] == "Friday"]
pusa_tip = df["tip"].loc[df["pickup_dayname"] == "Saturday"]
pus_tip = df["tip"].loc[df["pickup_dayname"] == "Sunday"]

```

```python
# ANOVA Test
f_stat, p_val = stats.f_oneway(pum_tip,put_tip,puw_tip,puth_tip,puf_tip,pusa_tip,pus_tip)
print("ANOVA F-statistic:", f_stat)
print("P-value:", p_val)

if p_val < 0.05:
    print("Reject null hypothesis:")
else:
    print("Fail to reject null hypothesis:")

```
